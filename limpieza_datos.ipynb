```python
columns_to_replace = ['PROMEDIO_NOTAS', 'PROMEDIO_CM_MAX', 'PROM_GRAL2020', 'PROM_GRAL2021', 'PROM_GRAL2022', 'PROM_GRAL2023']

# Replace commas with points and convert to numeric
for column in columns_to_replace:
    df[column] = df[column].replace(',', '.', regex=True).astype(float)
```

```python
df_categorical = df.select_dtypes(exclude=[np.number])
df_categorical.describe().T
```

```python
columns_to_check = ['PAGO_MATRICULA2020', 'PAGO_MENSUAL2020', 'PAGO_MATRICULA2021', 'PAGO_MENSUAL2021',
                    'PAGO_MATRICULA2022', 'PAGO_MENSUAL2022', 'PAGO_MATRICULA2023', 'PAGO_MENSUAL2023']

df2 = df.dropna(subset=columns_to_check)
```
