```python
columns_to_replace = ['PROMEDIO_NOTAS', 'PROMEDIO_CM_MAX', 'PROM_GRAL2020', 'PROM_GRAL2021', 'PROM_GRAL2022', 'PROM_GRAL2023']

# Replace commas with points and convert to numeric
for column in columns_to_replace:
    df[column] = df[column].replace(',', '.', regex=True).astype(float)
```

```python
# Determine the number of numerical columns
num_cols = df.select_dtypes(include='number').columns

# Set up subplots
num_plots = len(num_cols)
num_rows = (num_plots + 1) // 2  # Calculate number of rows, rounding up
fig, axes = plt.subplots(num_rows, 2, figsize=(12, 6*num_rows))

# Flatten axes array for easy indexing
axes = axes.flatten()

# Iterate over numerical columns and create box plots
for i, col in enumerate(num_cols):
    df.boxplot(column=col, ax=axes[i])
    axes[i].set_title(f'Box Plot for {col}')
    axes[i].set_ylabel('Value')
    axes[i].set_xlabel('')

# Remove empty subplots
if num_plots % 2 != 0:
    fig.delaxes(axes[-1])

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
columns_to_label_encode = ['NOMBRE_UNIDAD_EDUC', 'NOMBRE_COMUNA_EGRESO',
                            'PAGO_MATRICULA2020', 'PAGO_MENSUAL2020', 'PAGO_MATRICULA2021', 'PAGO_MENSUAL2021',
                            'PAGO_MATRICULA2022', 'PAGO_MENSUAL2022', 'PAGO_MATRICULA2023', 'PAGO_MENSUAL2023']


label_encoder = LabelEncoder()
for column in columns_to_label_encode:
    df3.loc[:, column] = label_encoder.fit_transform(df3.loc[:, column])
```

```python
df3.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df_normalized = pd.DataFrame(scaler.fit_transform(df3), columns=df3.columns)
```

```python
df_normalized.head()
```

```python
X = df_normalized.drop(['MRUN', 'CLEC_REG_ACTUAL', 'MATE1_REG_ACTUAL', 'MATE2_REG_ACTUAL', 'PROMEDIO_CM_MAX', ], axis=1)
```

```python
y = df_normalized['PROMEDIO_CM_MAX']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
linreg.fit(X_train, y_train)
```

```python
print("R squared of the Linear Regression on training set: {:.2%}".format(linreg.score(X_train, y_train)))
print("R squared of the Linear Regression on test set: {:.2%}".format(linreg.score(X_test, y_test)))
```

```python
y_pred = linreg.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_squared_log_error
print("Mean Squared Error on test set is {:.2%}".format(mean_squared_error(y_test, y_pred)))
```
