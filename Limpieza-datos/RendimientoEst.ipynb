```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("R2023.csv", delimiter=';')
df1 = pd.read_csv("R2022.csv", delimiter=';')
df2 = pd.read_csv("R2021.csv", delimiter=';')
df3 = pd.read_csv("R2020.csv", delimiter=';')
df_prueba= pd.read_csv("2024.csv", delimiter=';')
```

```python
# Supongo que df1, df2, df3 también son DataFrames que contienen la columna 'MRUN'
#df_prueba = df_prueba.to_frame()
#df_prueba = df_prueba['MRUN']

columnas2020 = []
columnas2021 = []
columnas2022 = []
columnas2023 = []

for each in df_prueba['MRUN']:
    # Para df
    row = df[df['MRUN'] == int(each)]
    if not row.empty:
        columnas2023.append([row['GEN_ALU'].values[0], row['EDAD_ALU'].values[0], row['COD_COM_ALU'].values[0], row['PROM_GRAL'].values[0], row['ASISTENCIA'].values[0], row['COD_JOR'].values[0], row['COD_COM_RBD'].values[0]])

    # Para df1
    row1 = df1[df1['MRUN'] == int(each)]
    if not row1.empty:
        columnas2022.append([row1['PROM_GRAL'].values[0], row1['ASISTENCIA'].values[0]])

    # Para df2
    row2 = df2[df2['MRUN'] == int(each)]
    if not row2.empty:
        columnas2021.append([row2['PROM_GRAL'].values[0], row2['ASISTENCIA'].values[0]])

    # Para df3
    row3 = df3[df3['MRUN'] == int(each)]
    if not row3.empty:
        columnas2020.append([row3['PROM_GRAL'].values[0], row3['ASISTENCIA'].values[0]])
```

```python
MRUN = pd.DataFrame(df_prueba['MRUN'])


df2020 = pd.DataFrame(columnas2020, columns=[ 'PROM_GRAL2020', 'ASISTENCIA2020'])
df2021 = pd.DataFrame(columnas2021, columns=[ 'PROM_GRAL2021', 'ASISTENCIA2021'])
df2022 = pd.DataFrame(columnas2022, columns=[ 'PROM_GRAL2022', 'ASISTENCIA2022'])
df2023 = pd.DataFrame(columnas2023, columns=['GEN_ALU2023', 'EDAD_ALU', 'COD_COM_ALU2023','PROM_GRAL2023', 'ASISTENCIA2023', 'COD_JOR', 'COD_COM_RBD'])

MRUN = pd.concat([MRUN.reset_index(drop=True), df2020, df2021, df2022, df2023], axis=1)

```

```python
MRUN['MismaCom'] = np.where(MRUN['COD_COM_RBD'] == MRUN['COD_COM_ALU2023'], 1, 0)
```

```python
MRUN.to_csv("rendimiento.csv",index=False)
```
